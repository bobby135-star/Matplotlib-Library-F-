## 12.STACKED BAR CHART WITH EXAMPLE IN MATPLOTLIB || HORIZONTAL & VERTICAL STACKED BAR CHART || PYTHON

```python
from matplotlib import pyplot as plt
import piplite
await piplite.install('matplotlib')
Years=['2017','2018','2019','2020']
Placements=[150,275,324,295]

cse=[75,80,47,82]
it=[46,58,35,52]
ece=[67,59,72,62]

ece_start=[cse[i]+it[i] for i in range (len(cse))]

plt.title("PLACEMENTS COMPARISION")
plt.bar(Years,cse,width=0.4,color="black")
plt.bar(Years,it,bottom=cse,width=0.4,color="green")
plt.bar(Years,ece,bottom=ece_start,width=0.4,color="red")
plt.xlabel("Year")
plt.ylabel("Placements")
plt.ylim(0,300)
plt.show()
```

```python
from matplotlib import pyplot as plt
import piplite
await piplite.install('matplotlib')
Years=['2017','2018','2019','2020']
Placements=[150,275,324,295]

cse=[75,80,47,82]
it=[46,58,35,52]
ece=[67,59,72,62]
eee=[53,47,36,52]

ece_start=[cse[i]+it[i] for i in range (len(cse))]
eee_start=[ece_start[i]+ece[i] for i in range (len(ece))]

plt.title("PLACEMENTS COMPARISION USING STACKED BAR GRAPH")
plt.barh(Years,cse,height=0.4,color="black",label="CSE")
plt.barh(Years,it,left=cse,height=0.4,color="green",label="IT")
plt.barh(Years,ece,left=ece_start,height=0.4,color="red",label="ECE")
plt.barh(Years,eee,left=eee_start,height=0.4,color="blue",label="EEE")

plt.xlabel("Year")
plt.ylabel("Placements")
plt.xlim(0,350)
plt.legend()

plt.show()
```

# 01.How to change the matplotlib figure size with figsize and rcParams || Matplotlib Tips
# Matplotlib: Changing the Figure Size





```python
from matplotlib import pyplot as plt
import piplite
import numpy as np
await piplite.install('matplotlib')

X = np.linspace(0, 4*np.pi, 1000)
Y = np.sin(X)
plt.plot(X, Y);
plt.show()
```

## figsize

Reference the figure and its `figsize` argument to update the figure size of an individual figure.

The tuple you pass will be read in as (width in inches, width in height).

```python
plt.figure(figsize=(10,6))
plt.plot(X, Y);
plt.show()
```

```python
plt.figure(figsize=(6, 10))
plt.plot(X, Y);
plt.show()
```

## via rcParams

Update your run command (rc) parameters to change the size of ALL future figures.

```python
plt.plot(X, Y);
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (4, 4)
```

```python
plt.plot(X, Y);
plt.show()
```

```python
plt.plot(X, -Y);
plt.show()
```

# 03. Add math symbols to matplotlib figures with matplotlib LaTeX || Matplotlib Tips

# Matplotlib: Adding Symbols with $\TeX$

```python
from matplotlib import pyplot as plt
import piplite
import matplotlib
import numpy as np
await piplite.install('matplotlib')
```

```python
X = np.linspace(1, 4, 1000)
Y = X
Y2 = X**2
Y_log = np.log(X)
Y_cos = np.cos(np.pi*X)
```

## Intro Figures

```python
matplotlib.rcParams['font.size'] = 16
plt.plot(X, Y2, X, Y, X, Y_log)
plt.xlabel('$x \in [1, \infty)$', fontsize=14)
plt.ylabel('$y$',fontsize=14)
plt.legend(['$y=x^2$', '$y=x$', '$y=\ln\;x$'], fontsize=12)
plt.title('Comparison of $\mathcal{O}(x^2)$, $\mathcal{O}(x)$, and $\mathcal{O}(\ln\;x)$', fontsize=16)
plt.text(3.4, 8, 'Rapid growth\nwhen $y \sim \mathcal{O}(x^2)$', fontsize=14, color='gray')

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False);
plt.show()
```

## Symbols and Sub/Superscripts

Write symbols and sub/superscripts in matplotlib with $\TeX$.

**!!!** Be sure to use the 'r' prefix so that Python leaves the text as a raw string and does not convert escape sequences.

**Superscripts**

```python
plt.plot(X, Y2)
plt.title(r'$y=x^2$');
plt.show()
```

**Subscripts**

```python
plt.plot(X, Y)
plt.title(r'$y=x_i$');
plt.show()
```

**Symbols: Greek letters, Math functions**

```python
plt.plot(X, Y_cos)
plt.text(2.3, 0.75, r'$y=\cos(\pi x)$');  
plt.show()
```

```python
plt.plot(X, Y_cos)
plt.title(r'$y=\cos(\pi x)$')
plt.text(2.3, 0.75, r'$y=\cos(\pi x)$')
plt.xlabel(r'$y=\cos(\pi x)$')
plt.ylabel(r'$y=\cos(\pi x)$')
plt.legend([r'$y=\cos(\pi x)$'], loc=4); 
plt.show()
```

## Fractions and Formulas

$\TeX$ also allows you to create fractions and full equations.

**!!!** Be sure to use the 'r' prefix so that Python leaves the text as a raw string and does not convert escape sequences.

```python
x = np.linspace(0.5, 100, 1000)
y_frac = 2 + 1/(x**2);
```

**Fractions**

```python
plt.plot(x, y_frac)
plt.ylim(0, None)
plt.title(r'$y = 2 + \frac{1}{x^2}$', fontsize=20); 
plt.show()
```

```python
plt.plot(x, y_frac)
plt.ylim(0, None)
plt.text(40, 4, r'$\lim_{x\rightarrow\infty}\;2 + \frac{1}{x^2} = 2$', fontsize=20);
plt.show()
```

## BONUS Content

**Accent Marks**

```python
plt.plot(X, Y_log)
plt.title(r'$\ddot{U}$ber cool results!');
plt.show()
```

**Equation Array**

```python
x_10 = np.linspace(10.5, 100, 1000)
y_frac_10 = 1 + 1/(x_10-10)**2
print(x_10)
print(y_frac_10)
```

Be sure to update your rc parameters to "usetex" in order to use equation arrays.

```python
from matplotlib import rc
rc('text', usetex=True)
```

```python
plt.plot(x, y_frac)
plt.plot(x_10, y_frac_10)
plt.ylim(0, None)
eq = (r'\begin{eqnarray*}'
      r'\lim_{x\rightarrow\infty}\;2 + \frac{1}{x^2}&=&2 \\'
      r'\lim_{x\rightarrow\infty}\;1 + \frac{1}{(x-10)^2}&=&1 '
      r'\end{eqnarray*}')
plt.text(40, 5, eq);
```
