# Matplotlib: Changing Axes Limits

```python
from matplotlib import pyplot as plt
import piplite
import numpy as np
await piplite.install('matplotlib')
```

```python
X = np.linspace(1, 4, 1000)
Y = 2 + X**2
```

## pyplot xlim and ylim

Pass values to `plt.xlim()` or `plt.ylim()` to adjust the x- or y-axis limits.

```python
plt.plot(X, Y)
plt.xlim(0, 5);
plt.show()
```

```python
plt.plot(X, Y)
plt.ylim(0, 20);
plt.show()
```

```python
plt.plot(X, Y)
plt.xlim(0, 5)
plt.ylim(0, 20);
plt.show()
```

### Adjusting only one limit

```python
plt.plot(X, Y)
plt.ylim(0, None);
plt.show()
```

```python
plt.plot(X, Y)
plt.ylim(bottom=0);  #top, left, right also work
plt.show()
```

## axis set_xlim and set_ylim

Alternatively: pass values to `set_xlim()` or `set_ylim()` to specific axes.

```python
plt.plot(X, Y)

ax=plt.gca()
ax.set_xlim(0, 5)
ax.set_ylim(0, None);
plt.show()
```

```python
type(ax)
```

### Subplots

```python
plt.subplots(1, 3)

plt.subplot(1, 3, 1)
plt.plot(X, Y)

ax2 = plt.subplot(1, 3, 2)
plt.scatter(X[::100], Y[::100])

#or you could put the xlim, ylim functions 
#   here to change this subplot!

plt.subplot(1, 3, 3)
plt.plot(X, -Y)

plt.tight_layout()

plt.xlim(0, 5)
plt.ylim(-20, 0)

ax2.set_xlim(-2, 6)
ax2.set_ylim(-10, 25);
plt.show()
```

## BONUS: Reversing the axis

```python
plt.plot(X, Y)
plt.xlim(4, 1);
plt.show()
```

```python
plt.plot(X, Y)
left, right = plt.xlim()
plt.xlim(right, left);
plt.show()
```

#### Cereal Example

```python
import pandas as pd
```

```python
url = 'https://bit.ly/2JRmGC2'        #Pull in cereal data from GitHub repo

df = pd.read_csv(url, na_values=-1)   #Null values in this dataset are denoted -1

df['sugars_per_cup'] = df.sugars/df.cups
df['fat_per_cup'] = df.fat/df.cups
```

```python
plt.figure(figsize=(9, 5))

plt.scatter(df['sugars_per_cup'], df['fat_per_cup'], 
            s=100, alpha=0.7, zorder=2, edgecolor='black')

plt.axhline(df['fat_per_cup'].median(), linestyle='--', lw=3, 
            zorder=1, color='xkcd:terra cotta')
plt.axvline(df['sugars_per_cup'].median(), linestyle='--', lw=3, 
            zorder=1, color='xkcd:terra cotta')

plt.title('Few cereals are low sugar, high fat\n', 
          fontsize=20,
          color='xkcd:reddy brown',
          fontweight='bold',
          loc='left'
         )

for spine in plt.gca().spines.values():
    spine.set_visible(False)
plt.xticks([])
plt.yticks([])
  
plt.annotate(f'Median Fat {df.fat_per_cup.median():.1f}g', 
             (19, df.fat_per_cup.median()-0.6), 
             fontsize=12,
             color='xkcd:terra cotta'
            );
plt.annotate(f'Median Sugar {df.sugars_per_cup.median():.1f}g', 
             (df.sugars_per_cup.median()+0.5, 8), 
             fontsize=12,
             color='xkcd:terra cotta'
            )

#Add rectangle
plt.gca().add_patch(Rectangle((0, 1.5), 10, 8,
                     color='xkcd:reddy brown', lw=3, fill=None))

#Annotate rectangular region
plt.annotate(f'Low Sugar,\nHigh Fat', 
             (0.5, 8), 
             fontsize=14,
             fontweight='bold',
             color='xkcd:reddy brown'
            )

plt.tight_layout()
plt.show()
#plt.savefig('06shapes_cereal_application.png', dpi=200);
```

```python
from matplotlib import pyplot as plt

import numpy as np
```

## Hatching on Barplots

```python
plt.bar(range(len(values)), values, hatch='/'); 
plt.show()
```

### Patterns

```python
plt.bar(range(len(values)), values, hatch='x');
plt.show()
```

```python
plt.bar(range(len(values)), values, hatch='*');
plt.show()
```

```python
patterns = ['', '\\', '']
bars = plt.bar(range(len(values)), values)

for bar, pattern in zip(bars, patterns):
    bar.set_hatch(pattern);
plt.show()
```

### Rotating tick labels for subplots

Remember `plt.xticks()` refers to the most recent subplot only.  Must reference twice to rotate both sets of x-ticks.

```python
plt.subplots(1, 2)

plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.xticks(x, xlabels, rotation=60) #first subplot

plt.subplot(1, 2, 2)
plt.scatter(x, y)
plt.xticks(x, xlabels, rotation=60); #second subplot
plt.show()
```

## Text in General Position or at Variable

### General Position: `transform=ax.transAxes`

```python
def add_mean_line(data):
    mu = np.mean(data)
    
    ax=plt.gca()
    ax.axvline(mu, color='black', lw=3, ls='--')
    
    ax.text(0.7, 0.9, f'mean={mu:.2f}',
            transform=ax.transAxes,   #transforms positions to range from (0,0) to (1,1)
            fontweight='bold', fontsize=12)
```

```python
values = np.random.normal(10, 2, size=100)

plt.hist(values, color='lightblue')
add_mean_line(values);
plt.show()
```

### Variable in a Loop

```python
students = ['A', 'B', 'C', 'D', 'E']
grades = [80, 42, 68, 91, 77]

plt.bar(students, grades, edgecolor='black', color='xkcd:light gray blue')

for i, g in enumerate(grades):
    plt.text(i, g-8, f'{g}', ha='center', fontsize=16)
    
plt.xlabel('Student')
plt.yticks([])
sns.despine(left=True)
plt.show()
```
